```python
#Importing all the necessary Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#import eli5

pd.set_option('display.max_columns', None)

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, recall_score, confusion_matrix
from sklearn import tree
import graphviz
from sklearn import metrics
from sklearn.metrics import classification_report
#from pdpbox import pdp, 

from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
#import scikitplot as skplt
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler
RANDOM_SEED = 42
```

```python
#Loading training and test data
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
cols = ['target', 'ID_code']
X = train_data.drop(cols,axis=1)
y = train_data['target']
```

```python
X.head()
```

```python
scaler = StandardScaler()
```

```python
tr = scaler.fit_transform(X)
```

```python
X_train, X_val, y_train, y_val = train_test_split(tr,y, random_state = 1)
```

```python
parameters= {
    'penalty': ['l1', 'elasticnet', 'l2','none'],
    'C': [0.1,2,3,4,5],
    'class_weight': ['dict', 'balanced'],
    'solver': ['newton_cg', 'liblinear', 'sag', 'saga'],
}
```

```python
lgr = LogisticRegression()
```

```python
grid_lgr = GridSearchCV(lgr, parameters, n_jobs=-1)
grid_lgr.fit(X_train,y_train)
yt_p = grid_lgr.predict(X_train)
yv_p = grid_lgr.predict(X_val)
```

```python
grid_lgr.best_params_
```

```python
grid_lgr.best_score_
```

```python
lgr = LogisticRegression(C=4, class_weight='dict', penalty='l2' , solver='sag')
lgr.fit(X_train,y_train)
yt_p = lgr.predict(X_train)
yv_p = lgr.predict(X_val) 

print('Training Accuracy', np.mean(yt_p==y_train))
print('Validation  Accuracy', np.mean(yv_p==y_val))
```

```python
tr_auc = roc_auc_score(y_train,yt_p)
vt_auc = roc_auc_score(y_val, yv_p)
```

```python
tr_auc
```

```python
vt_auc
```

```python
print(classification_report(y_train, yt_p))
```

```python
print(classification_report(y_val, yv_p))
```

```python
x = metrics.confusion_matrix(y_val, yv_p)
y = metrics.confusion_matrix(y_train, yt_p)
```

```python
sns.heatmap(x,square=True, annot=True, fmt='d', cbar=False,
            )
plt.xlabel('true label')
plt.ylabel('predicted label')
```
